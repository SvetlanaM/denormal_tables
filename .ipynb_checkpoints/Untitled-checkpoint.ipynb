```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('test.csv', delimiter=',')
df.reset_index(inplace=True)
df.head(3)
```

```python
listOfIds = [df["variants_ids"]]
```

```python
new_df = pd.DataFrame(columns=["id", "v_id"])
count = 0
for item in listOfIds:
    for i in range(0, 6):
        a = item[i]
        
        if type(a) != float:
            a = a.split()
            new_df["v_id"] = a
            new_df["id"] = count
            count += 1
```
